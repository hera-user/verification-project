 ### collect the free atomic Boolean variables

```python
def is_uninterp(t):
    return is_app(t) and t.decl().kind() == Z3_OP_UNINTERPRETED

def subterms(t):
    def sub(t, ids, terms):
        if t.get_id() in ids:
            return
        ids |= { t.get_id() }
        terms += [ t ]
        for s in t.children():
            sub(s, ids, terms)
        return terms
    return sub(t, set([]), [])

def get_bits(fml):
    return { t for t in subterms(fml) if is_uninterp(t) }
```
